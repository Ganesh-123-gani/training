```python
import numpy as np 
import pandas as pd

```

```python
df=pd.read_csv('student_data.txt')
student_data=pd.DataFrame(df)
print(df)
```

```python
df['total_gain_marks']=df['Math']+df['Science']+df['English']
print(df)
```

```python
df['percentage']=(df['total_gain_marks']/300)*100
print(df)
```

```python
df['mean_marks']=df[['Math','Science','English']].mean(axis=1)
print(df[['Name', 'Math', 'Science', 'English','mean_marks']])
```

```python
df['Math_result']=df['Math'].apply(lambda x: 'Pass' if x>35 else 'Fail')
df['Science_result']=df['Science'].apply(lambda x: 'Pass' if x>35 else 'Fail')
df['English_result']=df['English'].apply(lambda x: 'Pass' if x>35 else 'Fail')
print(df)
```

```python
count=df.value_counts()
print(count)
```

```python
greater_than_80=df[df['percentage']>60].shape[0]
less_than_50=df[df['percentage']<50].shape[0]
print(f'total students greater than 80% :{greater_than_80}')
print(f'total students less than 50% :{less_than_50}')
```

```python
filter=df[(df['Grade']=='A') & (df['Attendance (%)']>90)]
print(filter)
```

```python
age_count=df[df['Age']<16]
print(age_count)
```
